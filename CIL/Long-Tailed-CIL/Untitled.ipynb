```python
import os
import numpy as np

# 设置文件夹路径
# folder_path = 'G:/DG/PL_Reg/L-Reg_NeurIPS24-main/Long-Tailed-CIL_cifar100/vis'
folder_path = 'G:/DG/PL_Reg/L-Reg_NeurIPS24-main/Long-Tailed-CIL_cifar100/imagenet_subset/50base_11tasks/imagenet_subset_ltio_lucir_2stage_no_gs_fixd_0/figures'



# 用来存储加载的 NumPy 数组的列表
numpy_arrays = []

# 遍历文件夹
# count = 0
for filename in os.listdir(folder_path):
    # 只处理 .npy 文件
    if filename.endswith('.npy'):
        # 构造完整的文件路径
        file_path = os.path.join(folder_path, filename)
        print(file_path)
        
        # 加载 .npy 文件并添加到列表中
        numpy_array = np.load(file_path)
        numpy_arrays.append(numpy_array)

# 查看结果
for idx, arr in enumerate(numpy_arrays):
    print(f"Array {idx+1}:")
    print(arr)
```
